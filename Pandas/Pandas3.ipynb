```python
import numpy as np
import pandas as pd
```

### Finding Missing Data

```python
data = {
    'A': [1,2,np.nan, 4,5],
    'B': [np.nan, 2,3,4,5],
    'C': [1,2,3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any() # in all column
```

### Removing Missing Data

```python
# Removing nan works in row
df.dropna() # Remove all rows where NaN
```

```python
df
```

```python
df.dropna(thresh=3) # keeps where at least 3(threshold) non-null values
```

### Filling the missing Data

```python
df.fillna(0)
```

```python
values = {'A':0, 'B':100,'C':300,'D':400} #fills in column wise
df.fillna(value=values)
```

```python
df.fillna(df.mean())
```

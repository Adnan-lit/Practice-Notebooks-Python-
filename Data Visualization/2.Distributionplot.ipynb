```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
sns.displot(df['total_bill'])
```

```python
sns.displot(df['total_bill'], bins=50)
```

```python
sns.histplot(df['total_bill'], bins=50)
```

```python
sns.histplot(df['total_bill'], bins=50, kde=True)
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins=50, kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'], bins=50, kde=True)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'scatter')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'reg')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex')
```

```python
sns.pairplot(df, hue = 'size', palette='rainbow')
```

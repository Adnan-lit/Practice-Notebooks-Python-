### Matplotlib Basics

```python
x = np.linspace(0,5,11)
x
```

```python
y = x ** 2
y
```

```python
plt.plot(x,y)
plt.title("Hiiiii")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

### Subplots in Matplotlibs

```python
plt.subplot(2,2,1) #(row, col, select subplot)
```

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2) # To work on specific plot, work just under the subplot
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
```

### Object Oriented Way

```python
fig = plt.figure() #It's just a canvas
axis1 = fig.add_axes([0.1,0.1,0.8,0.8]) #[from left, from right, % in height, % in width]
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,0.3,0.3])
axis2.plot(x,y)

#Full control of canvas to plot
```

#### Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. I can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch)

```python
fig = plt.figure(figsize=(20,8), dpi=100)
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,0.3,0.3])
axis2.plot(x,y)
```

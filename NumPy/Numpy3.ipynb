### ARRAY OPERATIONS

```python
a1 = np.array([1,2,3,4,5])
a2 = np.array([6,7,8,9,10])
```

```python
a1 + a2
```

```python
a1 - a2
```

```python
a1 * a2
```

```python
a1 / a2
```

```python
a1 // a2
```

```python
a1 ** a2
```

### Deep and Shallow copy

```python
a = np.arange(1,21)
a
```

```python
slice = a[:5]
slice = slice * 10
slice
```

```python
a
```

```python
b = a #Deep copy
b = a.copy() #shallow copy
```

### Advance Array Manipulation

#### Stacking Array

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
```

```python
np.vstack((a,b))
```

```python
np.hstack((a,b))
```

```python
np.column_stack((a,b))
```

#### Splitting Array

```python
c = np.arange(16).reshape(4,4)
c
```

```python
np.hsplit(c,2)
```

```python
np.vsplit(c,2)
```

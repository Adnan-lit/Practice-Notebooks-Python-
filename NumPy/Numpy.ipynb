```python
import numpy as np
```

### Basic Array Creation

```python
arr = np.array([1,2,3,4])
arr
```

```python
a = [1,2,3.5, "hello"]
a
```

```python
np.array(a) #converts in string all element
```

```python
a = [1, 2, 3.5]
np.array(a) #converts in float all element
```

```python
l = [[1,2,3],[4,5,6],[7,8,9]]
l #list does not supports multi-dimension
```

```python
np.array(l) #numpy supports multi-dimension
```

### Array generation function

```python
np.arange(1,11,2) #1 to 10 steps 2
```

```python
arr = np.zeros((6,5)) #single value gives array ... for matrix(row,col)
arr
```

```python
arr = np.ones(6)
arr
```

```python
arr = np.linspace(1,5,2) #1to5 2elements linear space
arr
```

### Random Generation function

```python
np.random.rand(10) #vector
# value between 0 and 1
```

```python
np.random.randint(6) #0 to n
np.random.randint(10,20)  #10 to 20
np.random.randint(10,20,10)  #10 to 20 ==> 10 numbers
```

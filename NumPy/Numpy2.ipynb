```python
import numpy as np
```

### NUMPY INDEXING AND SLICING OF VECTORS

```python
arr = np.arange(11,21)
arr
```

```python
arr[3:6] # index x to n-1
arr[:6] # start to 6
arr[3:] #3 to end
arr[3::2] # 3to end steps = 2
```

### NUMPY INDEXING AND SLICING OF MATRIX

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
arr[0,0] # row,col
```

```python
slice = arr[3:,3:]
slice
```

```python
arr[:,2]
```

### BOOLEAN INDEXING

```python
arr = np.arange(11,21)
arr
```

```python
bool_index = arr%2 == 0
bool_index
```

```python
arr = arr[bool_index]
arr
```
